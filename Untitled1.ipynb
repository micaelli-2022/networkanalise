```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()

G.add_node('M')
G.add_node('N')
G.add_node('O')
G.add_node('P')
G.add_node('Q')
G.add_node('R')
G.add_node('S')

G.add_edge('M', 'S') 
G.add_edge('N', 'O')
G.add_edge('P', 'R')
G.add_edge('N', 'S')
G.add_edge('O', 'M')
G.add_edge('N', 'Q')
G.add_edge('O', 'M')
G.add_edge('P', 'P')
G.add_edge('S', 'M')
G.add_edge('O', 'N')
G.add_edge('S', 'M')
G.add_edge('N', 'R')
G.add_edge('P', 'M') 
G.add_edge('M', 'S')
```

```python
nx.draw(G,with_labels=True)
```
